```python
import os

base_path = "/content/drive/MyDrive/dataset"

folders = [
    "train/healthy",
    "train/diseased",
    "val/healthy",
    "val/diseased"
]

for folder in folders:
    os.makedirs(os.path.join(base_path, folder), exist_ok=True)

print("Folders created successfully!")
```

```python
train_dir = "/content/drive/MyDrive/dataset/train"
val_dir = "/content/drive/MyDrive/dataset/val"
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
```

```python
train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    zoom_range=0.3,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.5,1.5]
)

val_gen = ImageDataGenerator(rescale=1./255)
```

```python
IMG_SIZE = (224, 224)

train_data = train_gen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    class_mode='categorical'
)

val_data = val_gen.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    class_mode='categorical'
)
```

```python
base_model = MobileNetV2(weights='imagenet', include_top=False)
base_model.trainable = False

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(2, activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(train_data, validation_data=val_data, epochs=5)
```

```python
model.save("plant_model.h5")
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np

# Load model
model = load_model("plant_model.h5")

# Class labels (VERY IMPORTANT - adjust if needed)
class_names = ["diseased", "healthy"]

def predict(img_path):
    img = image.load_img(img_path, target_size=(224,224))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    prediction = model.predict(img_array)
    predicted_class = class_names[np.argmax(prediction)]

    return predicted_class

# Test with image
result = predict("leaf.jpeg")  # upload any image
print("Prediction:", result)
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np

# Load model
model = load_model("plant_model.h5")

class_names = ["healthy", "diseased"]  # adjust if needed

def predict_with_confidence(img_path):
    img = image.load_img(img_path, target_size=(224,224))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    prediction = model.predict(img_array)

    predicted_index = np.argmax(prediction)
    predicted_class = class_names[predicted_index]
    confidence = prediction[0][predicted_index] * 100

    return img, predicted_class, confidence

# Run prediction
img, label, conf = predict_with_confidence("leaf.jpeg")

# Show image
plt.imshow(img)
plt.axis('off')
plt.title(f"Prediction: {label} ({conf:.2f}%)")
plt.show()
```
